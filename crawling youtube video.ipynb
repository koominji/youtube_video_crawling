```python
#필요한 라이브러리 추가

from selenium import webdriver
from selenium.webdriver.support.ui import Select
import time
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
# 유튜브 금쪽이 플레이리스트 사이트

driver_path='/chromedriver.exe'

driver = webdriver.Chrome(executable_path=driver_path)
driver.implicitly_wait(3)
driver.get("https://www.youtube.com/watch?v=3zqf5oqkhsk&list=PLl9GPcxBUXInqV6GfDJqs1VYQWxquNZ8u")
driver.implicitly_wait(3)

time.sleep(0.5)
```

```python
# 페이지 코드 가져오기
page = driver.page_source
soup = BeautifulSoup(page,"html.parser") #html로 파싱

# 플레이리스트의 영상 링크
videos=soup.find_all('a',{'id':'thumbnail'})
```

```python
# 총 비디오 
len(videos)
```

```python
# 비디오 url -> youtube dowmloader 사이트 검색창에 넣어서 다운로드 

from selenium.webdriver.common.by import By


for video in videos:
    try:
        url="https://www.youtube.com" + video['href']
        print(url)
    
        # youtube downloader 사이트 열기
        driver.get("https://ssyoutube.com/en35/youtube-video-downloader")
        driver.implicitly_wait(3)

        time.sleep(3) 


        # 검색창 찾기 , 검색어 입력
        driver.find_element(By.ID,"id_url").click()
        time.sleep(3) 
        element=driver.find_element(By.ID,"id_url")
        time.sleep(3) 
        element.send_keys(url)

        # 검색 버튼 클릭
        driver.find_element(By.ID,"search").click()
        print("search_btn_clicked")
        time.sleep(3) 
        driver.find_element(By.ID,"download-mp4-720-audio").click()
        print("download_btn_clicked")
        time.sleep(5)
        
    except:
        print(video, "예외처리")
        

```
